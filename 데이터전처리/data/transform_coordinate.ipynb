```python
# Projection 정의
# UTM-K
proj_UTMK = Proj(init='epsg:5178') # UTM-K(Bassel) 도로명주소 지도 사용 중

# epsg:5181
proj_epsg_5181 = Proj(init='epsg:5181') # epsg:5181

proj_WGS84 = Proj(init='epsg:4326') # Wgs84 경도/위도, GPS사용 전지구 좌표
```

```python
953177.3125, 1953313
```

```python
# UTM-K -> WGS84 샘플
#x1, y1 = 953937.9, 1952051.9
x1, y1 = 953177.3125, 1953313
#x1, y1 = 753645.9623123633, 1501401.8033890715
x2, y2 = transform(proj_UTMK, proj_WGS84,x1,y1)
print(x2,y2)
```

```python
#x1, y1 = 126.9784147, 37.5666805
x1, y1 = 127.013626, 37.479434
x2, y2 = transform(proj_epsg_5181, proj_WGS84, x1, y1)
x3, y3 = transform(proj_WGS84, proj_UTMK, x2, y2)

print(x2,y2)
print(x3,y3)
```

```python
# epsg:5181 -> UTM-K 샘플
x1, y1 = 37.5666805, 126.9784147
x2, y2 = transform(proj_epsg_5181, proj_UTMK, x1, y1)
print(x2,y2)

# x, y 컬럼을 이용하여 UTM-K좌표를 WGS84로 변환한 Series데이터 반환
def transform_epsg_5181_to_utmk(df):
    return pd.Series(transform(proj_epsg_5181, proj_UTMK, df['x'], df['y']), index=['x', 'y'])

df_xy = pd.DataFrame([
                                        ['A', 961114.519726,1727112.269174],
                                        ['B', 940934.895125,1685175.196487],
                                        ['C', 1087922.228298,1761958.688262]
                                    ], columns=['id', 'x', 'y'])

df_xy[['x_w84', 'y_w84']] = df_xy.apply(transform_epsg_5181_to_utmk, axis=1)
```
