# 문체부 API 사용

## 크롤링

```python
res = []
for i in range(1, 13):
    # 문제부 API
    url = 'http://175.125.91.94/oasis/service/rest/meta4/getKCPG0504?numOfRows=100&pageNo='+str(i)
    a = requests.get(url)
    html = a.text
    
    soup = BeautifulSoup(html, 'html.parser')
    
    festival_info_url = soup.find_all("url")
    
    print('페이지 : ', i)
    
    # 문체부의 축제 정보 페이지로 넘어가서 자세한 정보 가져오기
    for num, festival in enumerate(festival_info_url):

        b = requests.get(festival.text)
        html2 = b.text

        soup2 = BeautifulSoup(html2, 'html.parser')
        
        if soup2.find('p', attrs={'class':'errorInfo'}):
            continue
        else:
            festival_info = dict()

            festival_name = soup2.find('div', attrs={'class': 'view_title color04'}).text
            festival_img = soup2.find('div', attrs={'class': 'culture_view_img'}).find('img')

            festival_info['title'] = festival_name
            festival_info['img'] = festival_img

            for dt, dd in zip(soup2.find_all('dt'), soup2.find_all('dd')):
                festival_info[dt.text] = dd.text.strip()

            festival_info['explanation'] = soup2.find('div', attrs={'class': 'view_con color04'}).text

            res.append(festival_info)
            print(num)
```

## 축제명 수정

- 축제명에 '년도' 들어간 경우 제거
- 띄어쓰기 제거

```python
festival_2015['축제명_수정'] = festival_2015['title'].str.replace(' 2015', '').str.replace('2015', '').str.replace('2015 ', '').str.replace(' ', '').str.replace('제\d\d회', '').str.replace('제\d회', '').str.strip()
festival_2016['축제명_수정'] = festival_2016['title'].str.replace(' 2016', '').str.replace('2016', '').str.replace('2016 ', '').str.replace(' ', '').str.replace('제\d\d회', '').str.replace('제\d회', '').str.replace('제\d\d\d돌', '').str.strip()
```

## 축제설명 수정

```python
festival_2015['축제설명_수정'] = festival_2015['explanation'].str.replace('\n', '').str.replace('\t', '')
festival_2016['축제설명_수정'] = festival_2016['explanation'].str.replace('\n', '').str.replace('\t', '')
```

## 문의 수정

```python
festival_2015['문의_수정'] = festival_2015['문의'].str.replace('\n', '').str.replace('\t', '')
festival_2016['문의_수정'] = festival_2016['문의'].str.replace('\n', '').str.replace('\t', '')
```

## 개최기간 수정

### ( ) 한글 등 없애기

```python
festival_2015['개최기간_수정'] = festival_2015['개최기간'].str.replace('\(금\)', '').str.replace('\(일\)', '').str.replace(' ', '')
festival_2016['개최기간_수정'] = festival_2016['개최기간'].str.replace('년', '.').str.replace('월', '.').str.replace('\d시', '').str.replace('[가-힣]', '').str.replace('\d\d:\d\d', '').str.replace('\(\)', '').str.replace('\( \)', '').str.replace('\/', '').str.replace(' ', '')
```

### 종료일/시작일로 나눠주기

```python
festival_2015 = pd.concat([festival_2015, festival_2015['개최기간_수정'].str.split('~', expand=True).rename(columns={0:'시작일', 1:'종료일'})], 1)
festival_2015['종료일'] = pd.DataFrame(np.where(festival_2015['종료일'].isnull(), festival_2015['시작일'], '2015.' + festival_2015['종료일'].astype(str))).astype('datetime64')
festival_2015['시작일'] = festival_2015['시작일'].astype('datetime64')
```

```python
festival_2016 = pd.concat([festival_2016, festival_2016['개최기간_수정'].str.split('~', expand=True).iloc[:, 0:2].rename(columns={0:'시작일', 1:'종료일'})], 1)

festival_2016['종료일'] = np.where((festival_2016['종료일'].isnull())|(festival_2016['종료일']==''), festival_2016['시작일'], festival_2016['종료일'])
festival_2016['시작일'] = festival_2016['시작일'].astype('datetime64')

tmp_ = []
for i, cell in enumerate(festival_2016['종료일']):
    if i in [5, 10, 23, 31, 52, 54, 57]:
        tmp_.append(str(festival_2016['시작일'][i].strftime('%Y.%m.'))+cell)
    elif '2016' not in cell:
        tmp_.append(str(festival_2016['시작일'][i].strftime('%Y.'))+cell)
    elif i==32:
        tmp_.append('2016.08.15')
    else:
        tmp_.append(cell)

festival_2016['종료일'] = pd.DataFrame(tmp_).astype('datetime64')
```

```python
url = "https://dapi.kakao.com/v2/local/search/keyword.json"
apikey = "7f6745340f5774fddc3560cf93bf54b3" 

result = list()
no_info_festival = list()

for nm, location in zip(festival_2015['축제명_수정'], festival_2015['축제장소_수정']):
    query = location #검색을 원하는 질의어 : 축제 장소

    r = requests.get( url, params = {'query':query, 'page':45, 'radius':200}, headers={'Authorization' : 'KakaoAK ' + apikey } ) 
    #결과를 json 형식으로 반환
    data = json.loads(r.text)
    print('축제명: {} \t\t 장소: {}'.format(nm, location))
    if data['meta']['total_count']!=0:        
        for i in data['documents']:
            address_name      = i['address_name']
            place_name        = i['place_name']
            place_url         = i['place_url']
            road_address_name = i['road_address_name']
            x                 = i['x']
            y                 = i['y']

            result.append({
                '축제명_수정'            : nm,
                '축제장소_수정'     : location,
                'address_name'      : address_name,
                'place_name'        : place_name,
                'place_url'         : place_url,
                'road_address_name' : road_address_name,
                'x'                 : x,
                'y'                 : y
            })
    else:
        result.append({
            '축제명_수정'            : nm,
            '축제장소_수정'     : location,
            'address_name'      : None,
            'place_name'        : None,
            'place_url'         : None,
            'road_address_name' : None,
            'x'                 : None,
            'y'                 : None
        })
        no_info_festival.append((nm, location))
```

## 기간지정

```python
driver = webdriver.Chrome('C:/Users/MINJI/chromedriver')
search_period = ['20190101', '20191231']
page_num = 1

driver.get('https://www.mcst.go.kr/kor/s_culture/festival/festivalList.jsp?pMenuCD=&pCurrentPage={}&pSearchType=01&pSearchWord=&pSeq=&pSido=01&pOrder=01down&pPeriod=&fromDt={}&toDt={}search_period'.format(page_num, search_period[0], search_period[1]))
tmp = driver.find_element_by_xpath('//*[@id="content"]/div[2]/div[3]/span')
p = re.compile('총 [\d]+건').match(tmp.text).group()
total_festival_num = int(re.findall('\d+', p )[0])

if total_festival_num%5 == 0: 
    last_page_num = int(total_festival_num/5)
else:
    last_page_num = int(total_festival_num/5)+1
    
    
res = []  
for page_num in range(1, last_page_num+1):
    print(page_num)
    driver.get('https://www.mcst.go.kr/kor/s_culture/festival/festivalList.jsp?pMenuCD=&pCurrentPage={}&pSearchType=01&pSearchWord=&pSeq=&pSido=01&pOrder=01down&pPeriod=&fromDt={}&toDt={}search_period'.format(page_num, search_period[0], search_period[1]))
    for i in range(1, 6):
        tmp = driver.find_element_by_xpath('//*[@id="content"]/div[2]/ul/li[{}]/a'.format(i))
        
        tmp.click()
          
        html2 = driver.page_source
        soup2 = BeautifulSoup(html2, 'html.parser')

        if soup2.find('p', attrs={'class':'errorInfo'}):
            continue
        else:
            festival_info = dict()

            festival_name = soup2.find('div', attrs={'class': 'view_title color04'}).text
            festival_img = soup2.find('div', attrs={'class': 'culture_view_img'}).find('img')

            festival_info['title'] = festival_name
            festival_info['img'] = festival_img

            for dt, dd in zip(soup2.find_all('dt'), soup2.find_all('dd')):
                festival_info[dt.text] = dd.text.strip()

            festival_info['explanation'] = soup2.find('div', attrs={'class': 'view_con color04'}).text

            res.append(festival_info)
            
        driver.get('https://www.mcst.go.kr/kor/s_culture/festival/festivalList.jsp?pMenuCD=&pCurrentPage={}&pSearchType=01&pSearchWord=&pSeq=&pSido=01&pOrder=01down&pPeriod=&fromDt={}&toDt={}search_period'.format(page_num, search_period[0], search_period[1]))
    page_num += 1
```
