# 시작일자 = 분석일자

- 분석의 용이성을 위해 기간을 하루로 특정
- 보통 축제 개최일에 가장 많은 사람이 오기때문에 시작일자를 분석일자로 정함

## 개최년월

```python
a['개최년월'] = a['시작일'].dt.strftime('%Y%m')
```

## 비교일자 추출

- 동년동기 : -14일, +14일 기간 중 같은 요일

```python
from datetime import datetime, timedelta

a['wk_1_before'] = a['시작일'] - timedelta(days=7)
a['wk_2_before'] = a['시작일'] - timedelta(days=14)

a['wk_1_after'] = a['시작일'] + timedelta(days=7)
a['wk_2_after'] = a['시작일'] + timedelta(days=14)
```

```python
a['분석일'] = a['시작일'].dt.strftime('%Y%m%d')
a['wk_1_before'] = a['wk_1_before'].dt.strftime('%Y%m%d')
a['wk_2_before'] = a['wk_2_before'].dt.strftime('%Y%m%d')
a['wk_1_after'] = a['wk_1_after'].dt.strftime('%Y%m%d')
a['wk_2_after'] = a['wk_2_after'].dt.strftime('%Y%m%d')
```

```python
a
```

```python
a = a.sort_values('분석일')
```

```python
a = a.reset_index(drop=True).reset_index().rename(columns={'index':'축제번호'})
a['축제번호'] = a['축제번호']+1
```

```python
for ym in a['개최년월'].unique():
    print(ym)
    f = open("query_{}.txt".format(ym), 'w')
    tmp = a.loc[a['개최년월'] == ym]
       
    for i in zip(tmp['축제번호'], tmp['wk_2_before'],  tmp['wk_1_before'],  tmp['분석일'],  tmp['wk_1_after'],  tmp['wk_2_after'], tmp['x_5178'], tmp['y_5178']):
        print(i)
        f.write("SELECT STD_YM, STD_YMD,\n")
        f.write("\t '{}' as festival_no, \n".format(i[0]))
        f.write("\t MAN_FLOW_POP_CNT_0004G + MAN_FLOW_POP_CNT_0509G + MAN_FLOW_POP_CNT_1014G + MAN_FLOW_POP_CNT_1519G as MAN_10,\n")
        f.write("\t MAN_FLOW_POP_CNT_2024G + MAN_FLOW_POP_CNT_2529G as MAN_20,\n")
        f.write("\t MAN_FLOW_POP_CNT_3034G + MAN_FLOW_POP_CNT_3539G as MAN_30,\n")
        f.write("\t MAN_FLOW_POP_CNT_4044G + MAN_FLOW_POP_CNT_4549G as MAN_40,\n")
        f.write("\t MAN_FLOW_POP_CNT_5054G + MAN_FLOW_POP_CNT_5559G as MAN_50,\n")
        f.write("\t MAN_FLOW_POP_CNT_6064G + MAN_FLOW_POP_CNT_6569G + MAN_FLOW_POP_CNT_70GU as MAN_60,\n")
        f.write("\t WMAN_FLOW_POP_CNT_0004G + WMAN_FLOW_POP_CNT_0509G + WMAN_FLOW_POP_CNT_1014G + WMAN_FLOW_POP_CNT_1519G as WMAN_10,\n")
        f.write("\t WMAN_FLOW_POP_CNT_2024G + WMAN_FLOW_POP_CNT_2529G as WMAN_20,\n")
        f.write("\t WMAN_FLOW_POP_CNT_3034G + WMAN_FLOW_POP_CNT_3539G as WMAN_30,\n")
        f.write("\t WMAN_FLOW_POP_CNT_4044G + WMAN_FLOW_POP_CNT_4549G as WMAN_40,\n")
        f.write("\t WMAN_FLOW_POP_CNT_5054G + WMAN_FLOW_POP_CNT_5559G as WMAN_50,\n")
        f.write("\t WMAN_FLOW_POP_CNT_6064G + WMAN_FLOW_POP_CNT_6570G + WMAN_FLOW_POP_CNT_70GU as WMAN_60\n")
        f.write("FROM TBDW_FLPOP_INFLOW_AGE_DAY_{}\n".format(ym))
        f.write("WHERE (STD_YMD = '{}' OR STD_YMD = '{}' OR STD_YMD = '{}' OR STD_YMD = '{}' OR STD_YMD = '{}')\n".format(i[1], i[2], i[3], i[4], i[5]))
        f.write("\t AND (X_COORD >= {0}-2 AND X_COORD <= {0}+2 \n".format(i[6]))
        f.write("\t AND Y_COORD >= {0}-10  AND Y_COORD <= {0}+10 )\n".format(i[7]))
        
        if tmp.shape[0] == 1:  
            continue
        else:
            f.write("UNION ALL \n")
    f.write(";")
    f.close()   
```
