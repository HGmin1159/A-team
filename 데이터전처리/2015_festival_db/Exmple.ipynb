기존 축제정보

주변 맛집정보

festival과 restaurants를 연결시켜주는 테이블

```python
fes_res = pd.read_csv("festival_restaurants.csv")
fes_res
```

예를들어 서울드럼페스티벌 축제 근처의 맛집 데이터:<br>
festival 데이터프레임에서 해당 id 찾기

```python
idx = festival["축제명"] == "서울드럼페스티벌"
festival_id = int(festival["id"][idx])
festival_id
```

res_fes 데이터프레임에서 restaurants_id 필터

```python
idx = fes_res["festival_id"] == festival_id
res_ids = list(fes_res["restaurants_id"][idx])
res_ids
```

restaurants 데이터프레임에서 필터

```python
idx = list(map(lambda x: x in res_ids, restaurants["id"]))
df = restaurants.loc[idx]
df
```
