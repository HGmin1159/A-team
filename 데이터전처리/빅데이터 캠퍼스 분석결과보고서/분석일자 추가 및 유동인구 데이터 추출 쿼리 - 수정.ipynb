## 비교일자가 201508 ~ 201607인 축제만

- 유동인구 데이터가 201508 ~ 201607 까지 있음

```python
a2 = a.loc[(a['wk_2_before']>=20150801)&(a['wk_2_after']<=20160731)]
```

```python
a2 = a2.drop(['축제번호'], 1).reset_index(drop=True).reset_index().rename(columns={'index':'축제번호'})
a2['축제번호'] = a2['축제번호']+1
```

```python
a2['비교년월_전_1'] = pd.to_datetime(a2['wk_1_before'], format='%Y%m%d').dt.strftime('%Y%m')
a2['비교년월_전_2'] = pd.to_datetime(a2['wk_2_before'], format='%Y%m%d').dt.strftime('%Y%m')

a2['비교년월_후_1'] = pd.to_datetime(a2['wk_1_after'], format='%Y%m%d').dt.strftime('%Y%m')
a2['비교년월_후_2'] = pd.to_datetime(a2['wk_2_after'], format='%Y%m%d').dt.strftime('%Y%m')
```
